## Eigen Values and Eigen Vectors

### Definition

In linear algebra, an eigenvector or characteristic vector of a linear transformation is a non-zero vector that does not change its direction when that linear transformation is applied to it. More formally, if T is a linear transformation from a vector space V over a field F into itself and v is a vector in V that is not the zero vector, then v is an eigenvector of T if T(v) is a scalar multiple of v. This condition can be written as the equation

T(v) = Lambda * v

T is nothing but another name for a square matrix. 

Let there be a square matrix, A

Mathemtically,

A is nxn matrix and v is nx1 vector.

Av = Lambda * v

Then a Lambda is called eigenvalue of A and for a given Lambda v is said to be an eigenvector for that lambda

## Properties

* When T acts on an eigenvector it can only get stretched but not rotated
* Can help understand a system, and can ease calculation in some places. eg M\*\*k \* x
* They are orthogonal, linearly independent and a set of eigenvectors for distinct eigenvalues forms a basis.

## Doing this by hand

Now,

Ax = Lambda * x

We can write this As,

(A - Lambda \* I) \* x = O

where I is identity matrix and O is zero matrix.

Take determinant on both side,

det((A - Lambda \* I) \* x = det(O)
det(A - Lambda \* I) * det(x) = o

X is non-zero,

so,

det(A - Lambda \* I) = 0 is one way of finding eigenvalues(Lambdas).

## Worked out example using SymPy

```python
import sympy as sm
from sympy.matrices import eye, Matrix

sm.init_printing()
```

```python
lam, a, b = sm.symbols('lambda a b')
lam, a, b
```

```python
I = eye(2)
I
```

```python
A = Matrix([[1, 2], [2, 1]])
A
```

```python
characteristic_eq = (A - lam * I).det()
characteristic_eq
```

```python
sm.solveset(characteristic_eq)
```

```python
lam1, lam2 = sm.solveset(characteristic_eq)
```

```python
x = Matrix([a, b])
x
```

```python
mat_eq = A*x - lam*x
mat_eq
```

```python
eq1, eq2 = mat_eq[0], mat_eq[1]
eq1, eq2
```

```python
eq1.subs(lam, lam1), eq2.subs(lam, lam2)
```

```python
sm.solve((eq1.subs(lam, lam1), eq2.subs(lam, lam1)), a, b)
```

```python
sm.solve((eq1.subs(lam, lam2), eq2.subs(lam, lam2)), a, b)
```
